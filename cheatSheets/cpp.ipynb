# C++ Syntax Cheat Sheet

## Variables and Data Types

### Variables
data_type variable_name = value;

### Data Types
- Integer: int
- Floating-point: float, double
- Character: char
- Boolean: bool

### Type Conversion
static_cast<data_type>(variable);

## Basic Operators
- Addition: +
- Subtraction: -
- Multiplication: *
- Division: /
- Modulus: %
- Increment: ++
- Decrement: --

## Strings
#include <string>
std::string my_string = "Hello, world!";

## Input and Output
#include <iostream>
std::cout << "Output: " << variable << std::endl;
std::cin >> variable;

## Control Flow
### If Statement
if (condition) {
    // code
} else if (condition) {
    // code
} else {
    // code
}

### Switch Statement
switch (expression) {
    case constant1:
        // code
        break;
    case constant2:
        // code
        break;
    default:
        // code
}

### While Loop
while (condition) {
    // code
}

### For Loop
for (initialization; condition; update) {
    // code
}

### Range-based For Loop
for (data_type variable : container) {
    // code
}

## Functions
### Defining a Function
return_type function_name(parameters) {
    // code
    return result;
}

### Calling a Function
function_name(arguments);

## Arrays
data_type array_name[array_size];

## Vectors
#include <vector>
std::vector<data_type> vector_name;

### Vector Methods
vector_name.push_back(x);
vector_name.pop_back();
vector_name.size();
vector_name.resize(new_size);
vector_name.clear();

## Pointers
data_type *pointer_name = &variable;

## References
data_type &reference_name = variable;

## Structs
struct StructName {
    data_type member1;
    data_type member2;
};

## Classes
class ClassName {
    public:
        // public members and methods
    private:
        // private members and methods
};

## Inheritance
class DerivedClass : public BaseClass {
    // derived class members and methods
};

## Error Handling
#include <stdexcept>
try {
    // code
} catch (const std::exception &e) {
    // code
}

## Namespaces
namespace NamespaceName {
    // code
}

### Using Namespaces
using namespace NamespaceName;

### Aliasing a Namespace
namespace Alias = NamespaceName;

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc

rc('text', usetex=True)

# LaTeX in axis-labels
matplotlib.rcParams.update({'font.size': 12, 'text.usetex': True})

x1, y1, x2, y2, vx1, vy1, vx2, vy2 = np.genfromtxt("Euler_dt00_1.csv", delimiter=',', unpack=True, skip_header=True)

# Create the plot
fig, axs = plt.subplot_mosaic("AB", figsize=(10, 6))

# Euler
ax1 = axs['A']
ax1.set_aspect('equal')
ax1.plot(x1, y1, label="$m_1=1$", lw=1)
ax1.plot(x2, y2, label="$m_2=2$", lw=1)
ax1.set_xlabel("$x$")
ax1.set_ylabel("$y$")
ax1.legend()
ax1.set_title("Euler method")

x1, y1, x2, y2, vx1, vy1, vx2, vy2 = np.genfromtxt("Verlet_dt00_1.csv", delimiter=',', unpack=True, skip_header=True)

# Verlet
ax2 = axs['B']
ax2.set_aspect('equal')
ax2.plot(x1, y1, label="$m_1=1$", lw=1)
ax2.plot(x2, y2, label="$m_2=2$", lw=1)
ax2.set_xlabel("$x$")
ax2.set_ylabel("$y$")
ax2.legend()
ax2.set_title("Verlet method")

fig.suptitle("Comparison of two integration methods - coarse step size $\Delta t = 0.1$")
fig.tight_layout()
fig.savefig("Euler_Verlet_a_coarse.pdf")
```

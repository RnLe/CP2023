```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc

rc('text', usetex=True)

# LaTeX in axis-labels
matplotlib.rcParams.update({'font.size': 12, 'text.usetex': True})

# Initialize the 1x3 subplot
fig, axs = plt.subplots(3, 1, figsize=(8, 15))

# List of data files
files = ["b)set_n4.dat", "b)set_n8.dat", "b)set_n16.dat"]
titles = ["Potential energy, kinetic energy and temperature $T(0)=1$ for N = 16 particles\nno prior equilibrium", "Potential energy, kinetic energy and temperature $T(0)=1$ for N = 64 particles\nno prior equilibrium", "Potential energy, kinetic energy and temperature $T(0)=1$ for N = 256 particles\nno prior equilibrium"]

# Iterate over each subplot and data file
for ax, file, title in zip(axs, files, titles):
    # Read the data from the file
    time, T, Ekin, Epot, vx, vy = np.genfromtxt(file, skip_header=True, delimiter=',', unpack=True)

    # Create the plot with multiple y-axes
    ax1 = ax
    ax2 = ax1.twinx()
    ax3 = ax1.twinx()

    # Move the spine of the third axis to the right
    ax3.spines['right'].set_position(('outward', 60))

    # Plot the data on the respective axes
    ax1.plot(time[2::10], T[2::10], 'x', label='T', color='tab:blue', markersize=3)
    ax2.plot(time[2:], Ekin[2:], label='Ekin', color='tab:orange')
    ax3.plot(time[2:], Epot[2:], label='Epot', color='tab:green')

    # Set plot labels and title
    ax1.set_xlabel(r'Time $\tilde{t}=t/\tau$')
    ax1.set_ylabel(r'T / $\frac{\epsilon}{k_B}$', color='tab:blue')
    ax2.set_ylabel(r'Ekin $/ \epsilon$', color='tab:orange')
    ax3.set_ylabel(r'Epot $/ \epsilon$', color='tab:green')
    ax.set_title(title)

    # Add a legend
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    ax3.legend(loc='center right')

# Adjust layout to avoid overlapping
plt.tight_layout()

# Save and show the plot
fig.savefig("b)sets.pdf")
plt.show()
```

```python
files = ["b)g_n4.dat", "b)g_n8.dat", "b)g_n16.dat"]
titles = ["Pair correlation function for $N = 16$, $T(0) = 1$\nno prior equilibrium", "Pair correlation function for $N = 64$, $T(0) = 1$\nno prior equilibrium", "Pair correlation function for $N = 256$, $T(0) = 1$\nno prior equilibrium"]

# Initialize the 1x3 subplot
fig, axs = plt.subplots(3, 1, figsize=(8, 15))

# Iterate over each subplot and data file
for ax, file, title in zip(axs, files, titles):
    # Read the data from the file
    rBin, g = np.genfromtxt(file, delimiter=",", skip_header=True, unpack=True)

    # Plot the data on the respective axes
    ax.plot(rBin, g)
    ax.set_xlim(0, (rBin[-1]//2) + 2)

    # Set plot labels and title
    ax.set_xlabel(r'Distance $\tilde{r}=r/\sigma$')
    ax.set_ylabel('Probability $g(r)$')
    ax.set_title(title)

# Adjust layout to avoid overlapping
plt.tight_layout()

# Save and show the plot
fig.savefig("b)g.pdf")
plt.show()
```

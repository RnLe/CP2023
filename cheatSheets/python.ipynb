# Python Syntax Cheat Sheet

## Variables and Data Types

### Variables
``variable_name = value``

### Data Types
- Integer: `int`
- Float: `float`
- String: `str`
- Boolean: `bool`

### Type Conversion
```python
int(x)
float(x)
str(x)
bool(x)
```

## Basic Operators
- Addition: `+`
- Subtraction: `-`
- Multiplication: `*`
- Division: `/`
- Modulus: `%`
- Exponentiation: `**`
- Floor division: `//`

## Strings
### Concatenation
``string1 + string2``

### Repetition
``string * n``

### Slicing
``string[start:end:step]``

## Lists
### Initialization
``my_list = [item1, item2, item3]``

### Accessing Elements
``my_list[index]``


### Slicing
``my_list[start:end:step]``


### List Methods
``my_list.append(x)``  
``my_list.extend(iterable)``  
``my_list.insert(index, x)``  
``my_list.remove(x)``  
``my_list.pop(index)``  
``my_list.index(x)``  
``my_list.count(x)``  
``my_list.sort()``  
``my_list.reverse()``  

## Control Flow
### If Statement
```python
if condition:
    # code
elif condition:
    # code
else:
    # code
```

### While Loop
```python
while condition:
    # code
```

### For Loop
```python
for variable in iterable:
    # code
```

## Functions
### Defining a Function
```python
def function_name(parameters):
    # code
    return result
```

### Calling a Function
``function_name(arguments)``


## Error Handling
### Try-Except Block
```python
try:
    # code
except ExceptionType:
    # code
```

### Raising Exceptions
```python
raise ExceptionType("Error message")
```

## Modules
### Importing a Module
```python
import module_name
```

### Importing Specific Items
```python
from module_name import item1, item2
```

### Aliasing a Module
```python
import module_name as alias
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# LaTeX in axis-labels
matplotlib.rcParams.update({'font.size': 12, 'text.usetex': True})

# Generate data to plot
S, R, Z, Psi = np.genfromtxt("rk4_all_r.txt", delimiter=',', skip_header=True, unpack=True)

# Mirror values at y axis
Z_ = np.concatenate((np.flip(Z), Z))
R_ = np.concatenate((np.flip(-R), R))

def find_solution_indices(R, threshold=0.5):
    indices = np.where(np.abs(R - threshold) < 4.5e-5)[0]
    return indices

def truncate_data(R, Z, index, start):
    return R[start+1:index + 1], Z[start+1:index + 1]

size_R_ = np.size(R_)
print(size_R_)

solution_indices = find_solution_indices(R)
R1, Z1 = truncate_data(R_, Z_, size_R_//2 + solution_indices[0], size_R_//2 - solution_indices[0])
R2, Z2 = truncate_data(R_, Z_, size_R_//2 + solution_indices[1], size_R_//2 - solution_indices[1])
print(Z1)
print(Z2)


# Create the plot
fig, axs = plt.subplot_mosaic("A;B")

# First drop
ax = axs['A']
ax.plot(R1, Z1, label="Shape of the liquid drop")
ax.vlines([-0.5, 0.5], [Z1.min(),Z1.min()], [Z1.max(),Z1.max()], linestyles='--', colors="tab:grey", label="$r=a/2$, edge of the tap")

# Set equal aspect ratio
ax.set_aspect('equal')
ax.grid()
ax.legend()
ax.set_xlabel("$R=r/a$")
ax.set_ylabel("$Z=z/a$")
ax.set_title("Dimensionless liquid drop model")

# Second drop with second solution to R = 1/2
ax2 = axs['B']
ax2.plot(R2, Z2, label="Shape of the liquid drop")
ax2.vlines([-0.5, 0.5], [Z2.min(),Z2.min()], [Z2.max(),Z2.max()], linestyles='--', colors="tab:grey", label="$r=a/2$, edge of the tap")

# Set equal aspect ratio
ax2.set_aspect('equal')
ax2.grid()
ax2.set_xlabel("$R=r/a$")
ax2.set_ylabel("$Z=z/a$")
ax2.set_title("Dimensionless liquid drop model - up to the second solution for $R=1/2$")

fig.tight_layout()

fig.savefig("c_r_2.pdf")

```

```python
# Generate data to plot
S, R, Z, Psi = np.genfromtxt("rk4_all_r.txt", delimiter=',', skip_header=True, unpack=True)

# Mirror values at y axis
Z_ = np.concatenate((np.flip(Z), Z))
R_ = np.concatenate((np.flip(-R), R))

# Create the plot
fig, ax = plt.subplots()

ax.plot(R_, Z_, label="Shape of the liquid drop")
ax.vlines([-0.5, 0.5], [Z_.min(),Z_.min()], [Z_.max(),Z_.max()], linestyles='--', colors="tab:grey", label="$r=a/2$, edge of the tap")

# Set equal aspect ratio
ax.set_xlim(-5,5)
ax.set_aspect('equal')
ax.grid()
ax.legend()
ax.set_xlabel("$R=r/a$")
ax.set_ylabel("$Z=z/a$")
ax.set_title("Dimensionless liquid drop model with multiple solutions")

fig.tight_layout()

fig.savefig("c_all_r.pdf")
```
